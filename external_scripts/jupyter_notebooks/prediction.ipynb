```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
import pickle
```

```python
filepath = 'prediction_data.csv'
X_pred = pd.read_csv(filepath)
X_pred.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
model_pkl = 'pickled_model.pkl'
with open(model_pkl, 'rb') as file_handle:
    model = pickle.load(file_handle)
    print(model)
```

```python
#aggregating dataset
X_pred_grouped = X_pred.groupby(['driver_id']).mean()
X_pred_grouped.head()
```

```python
danger_score = model.predict(X_pred_grouped)
```

```python
X_out = X_pred_grouped.copy()
X_out['safety_score'] = 1/danger_score
```

```python
X_out.head()
```

```python
filepath = 'SensifyPredictionData.csv'

X_out.to_csv(path_or_buf = filepath)
```
